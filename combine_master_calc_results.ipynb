```python
import csv

# File paths
file1 = 'LINES_FITS/Z_fit_212_r7_283_272.csv'
file2 = 'LINES_FITS/Z_fit_459_r2_152_157.csv'
output_file = 'LINES_FITS/combined.csv'

# Read the first file (including header)
with open(file1, 'r', newline='') as f1:
    reader1 = csv.reader(f1)
    data1 = list(reader1)

# Read the second file (skip header)
with open(file2, 'r', newline='') as f2:
    reader2 = csv.reader(f2)
    next(reader2)  # skip header
    data2 = list(reader2)

# Combine the data
combined_data = data1 + data2

# Write to output CSV
with open(output_file, 'w', newline='') as f_out:
    writer = csv.writer(f_out)
    writer.writerows(combined_data)

print(f"Files '{file1}' and '{file2}' have been combined into '{output_file}'.")
```

```python
import csv
import glob
import os

# Directory containing the CSV files
input_dir = "LINES_FITS"
output_file = os.path.join(input_dir, "combined.csv")

# Find all Z_fit CSV files
files = sorted(glob.glob(os.path.join(input_dir, "Z_fit_*.csv")))

if not files:
    raise RuntimeError("No Z_fit_*.csv files found.")

combined_data = []

for i, file in enumerate(files):
    with open(file, "r", newline="") as f:
        reader = csv.reader(f)
        rows = list(reader)

        if i == 0:
            # keep header from first file
            combined_data.extend(rows)
        else:
            # skip header for subsequent files
            combined_data.extend(rows[1:])

# Write combined CSV
with open(output_file, "w", newline="") as f_out:
    writer = csv.writer(f_out)
    writer.writerows(combined_data)

print(f"Combined {len(files)} files into '{output_file}'")
```
